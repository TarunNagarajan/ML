```python
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples = 200, factor = 0.4, noise = 0.05)
plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'bwr', edgecolor = 'k')
plt.title("non-linearly seperable concentric circles")
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis('equal')
plt.show()
```

Linear SVM attempt

```python
from sklearn.svm import SVC
import numpy as np 

linear = SVC(kernel='linear').fit(X, y)
plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'bwr', edgecolor = 'k')
plt.title("linear svm on circles")
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis('equal')

xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 200), np.linspace(-1.5, 1.5, 200))
Z = linear.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contour(xx, yy, Z, colors = 'k', levels = [0], linestyles = ['-'])
plt.show()
```

The Kernel Trick

```python
from sklearn.svm import SVC
import numpy as np 

rbf = SVC(kernel='rbf', gamma = 5).fit(X, y)
plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'bwr', edgecolor = 'k')
plt.title("rbf svm on circles")
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis('equal')

xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 200), np.linspace(-1.5, 1.5, 200))
Z = rbf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contour(xx, yy, Z, colors = 'k', levels = [0], linestyles = ['-'])
plt.show()
```
